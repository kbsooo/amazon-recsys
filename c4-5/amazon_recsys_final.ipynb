# Amazon GNN 추천 시스템 - 최종 통합 파이프라인

## 프로젝트 개요

**목표**: Amazon 사용자-아이템 구매 데이터를 활용한 GNN 기반 추천 시스템 구축

**핵심 특징**:
- 사용자별 맞춤 추천 개수 (기존 interaction의 50% 이하)
- Cold-start 사용자 (≤10개) 무조건 2개 추천
- 3가지 GNN 모델 비교: LightGCN, NGCF, SimpleGCN
- 다양한 평가 지표 및 시각화

**데이터셋**:
- 568,263 interactions
- 256,009 users
- 74,233 items
- Sparsity: 99.997%
